# **Importing Required Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **Loading the Dataset**

```python
data = pd.read_csv('/content/DoctorVisits - DA.csv')
```

# **Exploratory Data Analysis (EDA)**

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Get an overview of the dataset, including data types and non-null counts
print(data.info())
```

```python
# Summary statistics of numerical columns
print(data.describe())
```

### **Perform data analysis**

```python
#Count the number of visits by gender:
visits_by_gender = data['gender'].value_counts()
print(visits_by_gender)
```

```python
#Calculate the average age by gender:
average_age_by_gender = data.groupby('gender')['age'].mean()
print(average_age_by_gender)
```

```python
#Calculate the correlation between income and age:
correlation_income_age = data['income'].corr(data['age'])
print(correlation_income_age)
```

```python
healthcare_counts = data['private'].value_counts()
percentage_private = (healthcare_counts[1] / data.shape[0]) * 100
percentage_public = (healthcare_counts[0] / data.shape[0]) * 100
print(f"Private Healthcare: {percentage_private}%")
print(f"Public Healthcare: {percentage_public}%")
```

# **Data Visualization**

```python
# Gender distribution
gender_counts = data['gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```

```python
# Histogram for Age distribution
plt.hist(data['age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()
```

```python
# Income distribution
plt.hist(data['income'], bins=20)
plt.xlabel('Income')
plt.ylabel('Count')
plt.title('Income Distribution')
plt.show()
```

```python
# Visualization of illness types
illness_counts = data['illness'].value_counts()
plt.pie(illness_counts.values, labels=illness_counts.index, autopct='%1.1f%%')
plt.title('Illness Types')
plt.show()
```

```python
# Scatter plot between 'age' and 'income'
plt.scatter(data['age'], data['income'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Scatter Plot: Age vs. Income')
plt.show()
```

```python
# Horizontal bar graph for illness types
illness_counts = data['illness'].value_counts()
plt.barh(illness_counts.index, illness_counts.values)
plt.xlabel('Count')
plt.ylabel('Illness Type')
plt.title('Horizontal Bar Graph: Illness Types')
plt.show()
```

```python
# 2D Histogram for age and income
plt.hist2d(data['age'], data['income'], bins=(20, 20), cmap='Blues')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('2D Histogram: Age vs. Income')
plt.colorbar()
plt.show()
```

```python
# Box and whisker plot for 'income'
plt.boxplot(data['income'])
plt.ylabel('Income')
plt.title('Box and Whisker Plot: Income Distribution')
plt.show()
```

```python
# Box and whisker plot for 'age' grouped by 'gender'
data.boxplot(column='age', by='gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Box and Whisker Plot: Age by Gender')
plt.show()
```

```python
# Pie chart for private vs. non-private visits
private_counts = data['private'].value_counts()
plt.pie(private_counts.values, labels=private_counts.index, autopct='%1.1f%%')
plt.title('Pie Chart: Private vs. Non-Private Visits')
plt.show()
```

```python
# Grouped bar graph for 'gender' and 'private'
gender_private_counts = data.groupby('gender')['private'].value_counts().unstack()
gender_private_counts.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Grouped Bar Graph: Gender and Private Visits')
plt.legend(title='Private')
plt.show()
```
